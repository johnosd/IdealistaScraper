```python
# Ler o JSON e transformar em DataFrame
import json
import pandas as pd
caminho = "listaArrendar.json"
```

```python
with open(caminho, "r", encoding="utf-8") as f:
    dados = json.load(f)

df = pd.DataFrame(dados)
```

```python
df
```

```python
# Opcional: criar uma coluna numérica para "Contagem" (remove pontos e converte)
if "Contagem" in df.columns:
    contagem_num = (
        pd.to_numeric(
            df["Contagem"].astype(str).str.replace(".", "", regex=False),
            errors="coerce"
        ).astype("Int64")
    )
    df.insert(df.columns.get_loc("Contagem") + 1, "Contagem_num", contagem_num)

```
